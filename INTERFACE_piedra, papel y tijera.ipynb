```python
import pygame
import random
import sys
pygame.init()

# --- Pantalla ---
WIDTH, HEIGHT = 640, 480
screen = pygame.display.set_mode((WIDTH, HEIGHT))
pygame.display.set_caption("Piedra, Papel o Tijera")

# --- Colores y fuente ---
WHITE = (255, 255, 255)
BLACK = (0, 0, 0)
GRAY = (200, 200, 200)
GREEN = (0, 150, 0)
RED = (200, 0, 0)
font = pygame.font.SysFont(None, 36)
small_font = pygame.font.SysFont(None, 28)

# --- Cargar imágenes ---
imagenes = {
    "piedra": pygame.transform.scale(pygame.image.load("piedra.png"), (120, 120)),
    "papel": pygame.transform.scale(pygame.image.load("papel.png"), (120, 120)),
    "tijera": pygame.transform.scale(pygame.image.load("tijera.png"), (120, 120)),
}
# --- Opciones del juego ---
lista_opciones = ["piedra", "papel", "tijera"]
puntos_jugador1 = 0
puntos_ordenador = 0

# --- Botones de juego ---
botones = {
    "piedra": pygame.Rect(60, 330, 120, 120),
    "papel": pygame.Rect(260, 330, 120, 120),
    "tijera": pygame.Rect(460, 330, 120, 120)
}
# --- Variables del juego ---
jugador1 = ""
escribiendo_nombre = True
input_box = pygame.Rect(150, 200, 340, 40)
texto_nombre = ""
mensaje = ""
eleccion_jugador1 = ""
eleccion_ordenador = ""

# --- Botón jugar ---
boton_jugar = pygame.Rect(WIDTH // 2 - 75, 270, 150, 50)

# --- Lógica para decidir ganador ---
def determinar_ganador(j1, comp):
    global puntos_jugador1, puntos_ordenador
    if j1 == comp:
        return "Empate. ¡Inténtalo de nuevo!"
    elif (j1 == "piedra" and comp == "tijera") or \
         (j1 == "papel" and comp == "piedra") or \
         (j1 == "tijera" and comp == "papel"):
        puntos_jugador1 += 1
        return f"¡Enhorabuena {jugador1}! Ganas esta ronda."
    else:
        puntos_ordenador += 1
        return "¡Enhorabuena ordenador! Ganas esta ronda."
    
# --- Bucle principal ---
running = True
while running:
    screen.fill(WHITE)
    eventos = pygame.event.get()
    for event in eventos:
        if event.type == pygame.QUIT:
            running = False
        if escribiendo_nombre:
            if event.type == pygame.KEYDOWN:
                if event.key == pygame.K_RETURN and texto_nombre.strip() != "":
                    jugador1 = texto_nombre.strip()
                    escribiendo_nombre = False
                elif event.key == pygame.K_BACKSPACE:
                    texto_nombre = texto_nombre[:-1]
                else:
                    texto_nombre += event.unicode
            if event.type == pygame.MOUSEBUTTONDOWN:
                if boton_jugar.collidepoint(event.pos) and texto_nombre.strip() != "":
                    jugador1 = texto_nombre.strip()
                    escribiendo_nombre = False
        else:
            if event.type == pygame.MOUSEBUTTONDOWN and puntos_jugador1 < 3 and puntos_ordenador < 3:
                pos = pygame.mouse.get_pos()
                for opcion, rect in botones.items():
                    if rect.collidepoint(pos):
                        eleccion_jugador1 = opcion
                        eleccion_ordenador = random.choice(lista_opciones)
                        mensaje = determinar_ganador(eleccion_jugador1, eleccion_ordenador)
    if escribiendo_nombre:
        
        # Pantalla de entrada de nombre
        pygame.draw.rect(screen, GRAY, input_box, 2)
        txt_surface = font.render(texto_nombre, True, BLACK)
        screen.blit(txt_surface, (input_box.x + 10, input_box.y + 5))
        prompt = font.render("¿Cómo te llamas?", True, BLACK)
        screen.blit(prompt, (input_box.x, input_box.y - 40))
        if texto_nombre.strip() != "":
            pygame.draw.rect(screen, GREEN, boton_jugar)
            jugar_text = font.render("Jugar", True, WHITE)
            screen.blit(jugar_text, (boton_jugar.x + 35, boton_jugar.y + 10))
    else:
        # Pantalla del juego
        score_text = font.render(f"{jugador1}: {puntos_jugador1}  -  Ordenador: {puntos_ordenador}", True, BLACK)
        screen.blit(score_text, (WIDTH // 2 - score_text.get_width() // 2, 20))
        # Dibujar botones con imágenes
        for opcion, rect in botones.items():
            screen.blit(imagenes[opcion], rect.topleft)
        # Mostrar elecciones
        if eleccion_jugador1:
            e1 = small_font.render(f"{jugador1} eligió: {eleccion_jugador1}", True, BLACK)
            e2 = small_font.render(f"Ordenador eligió: {eleccion_ordenador}", True, BLACK)
            resultado = small_font.render(mensaje, True, BLACK)
            screen.blit(e1, (50, 170))
            screen.blit(e2, (50, 200))
            screen.blit(resultado, (50, 230))
        # Mostrar mensaje final
        if puntos_jugador1 == 3:
            final = font.render(f"¡{jugador1} gana la partida!", True, GREEN)
            screen.blit(final, (WIDTH // 2 - final.get_width() // 2, 270))
        elif puntos_ordenador == 3:
            final = font.render("¡Ordenador gana la partida!", True, RED)
            screen.blit(final, (WIDTH // 2 - final.get_width() // 2, 270))
    pygame.display.flip()
pygame.quit()
sys.exit()
```
